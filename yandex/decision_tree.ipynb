```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('data.csv')
df
```

```python
train_data = df[:12]
test_data = df[12:]
test_data
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(train_data[['Лекции','Семинары']], train_data['Зачет'])

test_data['Зачет'] = tree_model.predict(test_data[['Лекции','Семинары']])
test_data
```

```python
import sys


def main():
    """
    Пример ввода и вывода числа n, где -10^9 < n < 10^9:
    n = int(input())
    print(n)
    """
    n, costs = int(input()), list(map(int,input().split()))

    X = int(input())
    combo_products = map(int,input().split())

    k, desired_products = int(input()), list(map(int, input().split()))

    opt, dif = 0, 0
    for prod in desired_products:
        opt +=costs[prod]
        if prod in combo_products:dif+=costs[prod]
    print(opt - dif + X if dif)

if __name__ == '__main__':
    main()
```
