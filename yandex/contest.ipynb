```python
from collections import Counter
```

```python
n, m, q = 2, 5, 10
A, B = dict(Counter([1,2])), dict(Counter([1,2,3,4,5]))
queries = ['1 A 6', '1 A 4', '1 A 5', '1 A 6', '1 A 7', '-1 A 1', '1 B 7', '-1 A 6', '-1 B 1','1 A 7']
```

```python
r = 0
```

```python
for query in queries:
    query = query.split()
    type_card, player, card = int(query[0]), str(query[1]), int(query[2])
    if player == 'A':A[card] += type_card
    else:B[card]+=type_card
    X = set.union(A.keys(), B.keys())
    r+=abs(A[X] - B[X])
print(r)
```
